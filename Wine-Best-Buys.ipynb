```python
import pandas as pd
```

```python
wine_frame = pd.read_csv('./wine-reviews/wine_data_2.csv')
wine_frame =wine_frame.drop_duplicates(subset='title', keep='first', inplace=False)
```

```python
# Seeing what shape the data is in.
wine_frame[:5]
```

```python
# ****** California Cabernets ******
```

```python
# Create a slice that only Contains Califorina Cabernet Savignon
cali_cab =wine_frame[ wine_frame.province == 'California']
cali_cab =cali_cab[cali_cab.variety =='Cabernet Sauvignon']
```

```python
#How many Califorina Cabernet are in the Data Set
cali_cab_count=cali_cab.country.count()
cali_cab_count 
```

```python
# What is the average rating?
cali_cab.points.mean()
```

```python
# What is the average price?
cali_cab.price.mean()
```

### There are a total of 5317 Californa Cabernets in our data set, with an average rating of 89.1 and and Average Price of $59.60

This is a change of the original data of: 
There are a total of 7452 Californa Cab in our data set, with an average rating of 88.49 and and Average Price of $52.39
* the change was made after removing duplicated data

```python
# For every California Cab in the Data set, 
# Returns the Rating,
# returns the average price based on Rating, 
# along with the total number of wines with that rating
# and the standerd deveation based on price

cali_cab_avgs = []
for x in range (80,101):
    temp_slice =cali_cab[cali_cab.points == x]
    mean=temp_slice.price.mean()
    count=temp_slice.price.count()
    std=temp_slice.loc[:,"price"].std()
    cali_cab_avgs.append((x,mean,count,std))
    
cali_cab_avgs  
```

## Best Buys  --  Wines that are below  average price, at  each rating

```python
best_buys=[]
for x in range (20):
    temp=(cali_cab[(cali_cab.points ==cali_cab_avgs[x][0]) & (cali_cab.price <=(cali_cab_avgs[x][1]-cali_cab_avgs[x][3]))][['title','price','points']].sort_values(by='price'))
    if len(temp)>0:
        price=round(cali_cab_avgs[x][1],2)
        best_buys.append('At ' + str(cali_cab_avgs[x][0]) +' points, the average price is '+ str(price))
        best_buys.append(temp)
    else:
        best_buys.append('There are no wines that Qualify at ' + str(cali_cab_avgs[x][0]) +' points')
    best_buys.append('************************************************' )
best_buys
```
